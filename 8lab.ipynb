```python
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from matplotlib import pyplot as plt

pd.options.display.mpl_style = 'default'
spat = pd.read_csv('https://raw.githubusercontent.com/TeachingDataScience/data-science-course/forstudentviewing/08_data_results/data/msleep.csv')

```

```python


norm = pd.DataFrame({'d': np.random.normal(size=100)})
n_bins = np.abs(spat['sleep_rem'].max() - spat['sleep_rem'].min())
fig, axes = plt.subplots(nrows=2, ncols=2,)

norm['d'].plot(ax=axes[0, 0], kind='kde')
norm['d'].hist(ax=axes[0, 1], bins=n_bins/2)
spat['sleep_rem'].plot(ax=axes[1, 0], kind='kde')
spat['sleep_rem'].hist(ax=axes[1, 1], bins=n_bins/2)

print fig


```
